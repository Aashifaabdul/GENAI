```python
from langchain.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import   RecursiveCharacterTextSplitter
from langchain.embeddings import OllamaEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.chains.qa_with_sources.loading import load_qa_with_sources_chain
from langchain.llms import Ollama
import langchain
```

```python
llm= Ollama(model="mistral", temperature=0.7)
loader= UnstructuredURLLoader(urls=['https://www.pcworld.com/article/2686276/nvidias-geforce-rtx-5060-series-brings-dlss-4s-magic-to-the-masses.html',
                                    'https://www.pcworld.com/article/2686203/an-army-of-nvidias-affordable-geforce-rtx-5060-laptops-are-coming-soon.html'
])
data=loader.load()
len(data)
```

```python

text_splitter=RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap=200
)
docs=text_splitter.split_documents(data)
len(docs)
```

```python
docs[0].page_content
```

```python
embeddings = OllamaEmbeddings(model="nomic-embed-text")
```

```python
vector_index=FAISS.from_documents(docs,embeddings)
```

```python
vector_index
```

```python
chain=RetrievalQAWithSourcesChain.from_llm(llm=llm,retriever=vector_index.as_retriever())
chain
```

```python
query="what is the price of GeForce RTX 5060 Ti?"
```

```python
langchain.debug=True
chain({"question":query},return_only_outputs=True)
```
